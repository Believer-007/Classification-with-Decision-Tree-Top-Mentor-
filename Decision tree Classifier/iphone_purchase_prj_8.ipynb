# Importing Libraries and Modules

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

# Exploaratory Data Analysis

```python
data = pd.read_csv('iphone_purchase_records.csv')
data.head()
```

```python
shape = data.shape
print("There are {} rows and {} columns in the data".format(shape[0] , shape[1]))
```

```python
data.isnull().sum()
print("There are no any missing values in the data")
```

## Visualisation

```python
sns.barplot(data['Gender'])
plt.title("Gender")
plt.savefig('Graphs/gender.png')
plt.show()
```

```python
sns.boxplot(data['Salary'])
plt.title("Salary Distribution")
plt.savefig('Graphs/salary.png')
plt.show()
```

```python
sns.displot(data['Age'])
plt.title("Age")
plt.savefig('Graphs/Age.png')
plt.show()
```

## Data Transformation

```python
mapping = {'Male' : 0 , 'Female' : 1}
```

```python
data['Gender'] = data['Gender'].map(mapping)
data
```

# Data Splittting

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
X_train , X_test ,y_train , y_test = train_test_split(X,y,test_size= 0.25 , random_state = 0)
```

# Model Building

```python
model = DecisionTreeClassifier()
model.fit(X_train , y_train)
print("The model is fitted on training data")
```

# Model Evaluation

```python
y_pred = model.predict(X_test)
```

```python
score = accuracy_score(y_test , y_pred)
print(f"The score of the model is {score} ")
```

```python
matrix = confusion_matrix(y_test , y_pred)
print(f"The Confusion matrix of the model is \n {matrix} ")
```

```python
report = classification_report(y_test , y_pred)
print(f"The Classification Report of the model is \n {report} ")
```
